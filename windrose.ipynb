```python
import windrose as windroseAxes
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
Jerry=pd.read_csv('windrose data.csv')
Jerry
```

```python
ws=Jerry['WS10M']
ws
```

```python
wd=Jerry['WD10M']
wd
```

```python
from windrose import WindroseAxes

ax = WindroseAxes.from_ax()
ax.bar(wd,ws, normed=True, opening=0.9, edgecolor="white", cmap=plt.get_cmap('jet'),bins=np.arange(0.01, 8, 1))
ax.set_legend(title='wind speed(m/s)',fontsize=10, loc='upper right',bbox_to_anchor=(1.20,0.9))
ax.tick_params(axis='both', which='major',labelsize=10)
ax.set_title('WINDROSE FOR TONGO,TALENSI,UPPER EAST,GHA')
```
